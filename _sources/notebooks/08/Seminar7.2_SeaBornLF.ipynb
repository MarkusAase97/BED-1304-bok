# Oppgave 1a
I cellen over er datarammen `df` lest inn, som inneholder samme datasett som vi s친 p친 i seminaret for **matplotlib**.

Visualiser distribusjonen av karakterer i datasettet ved hjelp av et historgram fra SeaBorn.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Anta at datasettet er lastet som df
sns.histplot(df['Exam_Score'], kde=True)
plt.title('Distribusjon av Karakterer')
plt.xlabel('Poengsum p친 eksamen')
plt.ylabel('Frekvens')
plt.show()
```

# Oppgave 1b
Bruk Seaborn til 친 lage det samme plottet som over, men n친 skal du bare plottet de som har f친tt **karaktersum 80 eller h칮yere**.

```python
# Lager en kopi av dataframen v친r, og lager en ny kolonne
df_80 = df[df['Exam_Score'] >= 80]

# Samme som over
sns.histplot(df_80['Exam_Score'], kde=True)
plt.title('Distribusjon av Karakterer, for de over eller lik 80')
plt.xlabel('Poengsum p친 eksamen')
plt.ylabel('Frekvens')
plt.show()
```

# Oppgave 1c

Lag et "korrelasjonsvarmekart" (bedre kalt: **heatmap**) som viser hvordan ulike variabler i datasettet henger sammen.

## Visualiseringer
Det kan v칝re interessant 친 se p친 hvor mange bilder det er fra de ulike merkene, i ulike aldre.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Omformer datasettet 
df_long = df_bil.melt(id_vars=['Tid', 'Bilmerker'], 
                      value_vars=['Under 4 친r', '4 - 7 친r', '8 - 11 친r', '12 - 15 친r', '16 - 20 친r', 'Over 20 친r',], 
                      var_name='Aldersgruppe', 
                      value_name='Antall_biler')

# Visualisering med Seaborn
plt.figure(figsize=(12, 6))
sns.barplot(data=df_long, x='Bilmerker', y='Antall_biler', hue='Aldersgruppe', ci=None)
plt.xticks(rotation=90)
plt.title('Fordeling av biler i ulike aldersgrupper per bilmerke', fontsize=16)
plt.ylabel('Antall biler', fontsize=12)
plt.xlabel('Bilmerker', fontsize=12)
plt.tight_layout()
plt.show()
```

## Mer visualiseringer
Vi velger 친 se p친 bilmerker med h칮yest prosentandel vrakede biler, og i synkende rekkef칮lge - da blir det systematisk og fint.

```python
# Sorter bilmerkene etter prosentandel vrakede biler
df_sorted = df_bil.sort_values(by='Personbiler vraket mot pant (prosent av bestanden)', ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(data=df_sorted, x='Bilmerker', y='Personbiler vraket mot pant (prosent av bestanden)', palette='viridis')
plt.xticks(rotation=90)
plt.title('Bilmerker med h칮yest prosentandel vrakede biler', fontsize=16)
plt.ylabel('Prosentandel vrakede biler', fontsize=12)
plt.xlabel('Bilmerker', fontsize=12)
plt.tight_layout()
plt.show()
```

## Heatmap
Vi velger 친 se p친 korrelasjon mellom numeriske variabler

```python
# Velg numeriske kolonner for korrelasjon
corr_columns = ['Gjennomsnittsalder ved vraking (친r)', 
                'Personbilbestand 31. desember', 
                'Personbiler vraket mot pant', 
                'Personbiler vraket mot pant (prosent av bestanden)']

# Beregn korrelasjonen
corr_matrix = df_bil[corr_columns].corr()

# Plot korrelasjonsmatrise med Seaborn
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Korrelasjon mellom numeriske variabler', fontsize=16)
plt.tight_layout()
plt.show()
```
